## Run Application
Let's run an interactive python application

```python
!docker run -it python:3.12
```

### Run Code Inside Container

```python
!docker run --rm -v ${PWD}:/app -w /app python:3.12 python app.py

# Command Explanation
# docker run : strat a new container from an image
# --rm : delete container when finished (exit)

# Bind mount
# ${PWD} : current director on host machine
# /app : directory inside the container

# -w /app : set working directory inside container (i.e. cd /app)
# python:3:12 : image container is running
# python app.py : command executed inside container
```
